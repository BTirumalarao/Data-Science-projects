```python
#import the modules
import matplotlib.pyplot as plt
```

```python
#------------ Project Description---------------

#This project visualizes the temperature changes over a week using a line graph. 
#It helps you understand how to plot data points, add labels, titles, markers, and use Matplotlib to 
#show trends in weather conditions.
```

```python
# Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [30, 32, 31, 29, 28, 27, 26]
```

```python
plt.plot(days,temperature, ls='solid',c='blue', marker='o',mec='r',mfc='y')
plt.grid(axis='both')
plt.title("Weekly Temperature Report",c='g')
plt.xlabel('Days',c='r')
plt.ylabel('Temperature (Â°C)',c='r')
plt.show()
```
