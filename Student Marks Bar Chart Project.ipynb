```python
#import the modules
import matplotlib.pyplot as plt
```

```python
#------------ Project Description---------------

# This project displays studentsâ€™ marks in different subjects using a bar chart.
# It helps you learn how to create bars, set colors, add labels, and show values on 
# each bar to clearly compare performance across subjects.
```

```python
# Sample data
subjects = ['Math', 'Science', 'English', 'Computer', 'Telugu']
marks = [88, 75, 92, 95, 70]
```

```python
plt.bar(subjects,marks,color='b')
plt.title("Student Marks Chart",c='g')
plt.xlabel("subjects",c='r')
plt.ylabel("marks",c='r')
plt.ylim(0, 100)
plt.show()
```
